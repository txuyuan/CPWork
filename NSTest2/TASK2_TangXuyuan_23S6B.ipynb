```python
# Tang Xuyuan

# Task 2.1

def quickSortInts(arr, low, high):
    if high-low < 1:
        return
    pivot = low
    pivotValue = arr[pivot]
    left, right = low+1, high
    
    while left <= right:
        # Advance left to first >pivot value
        while (left <= high) and (arr[left] <= pivotValue):
            left += 1
        # Advance right to first <pivot value
        while (right > low) and (arr[right] >= pivotValue):
            right -= 1
        if left <= right:
            arr[left], arr[right] = arr[right], arr[left]
            
    arr[low], arr[right] = arr[right], arr[low]
    
    quickSortInts(arr, low, right-1)
    quickSortInts(arr, right+1, high)
        
        
        
# Read integers
file = open("INTEGERS.txt", "r")
integers = []
for line in file:
    line = line.strip()
    if line == "":
        continue
    integers.append(int(line))
file.close()

# Sort integers
quickSortInts(integers, 0, len(integers)-1)

# Write sorted integers
file = open("SORTED_TangXuyuan_NUM_INDEX.txt", "w")
for num in integers:
    file.write(str(num) + "\n")
file.close()

print(integers)
```

```python
# Tang Xuyuan

# Task 2.2

def BinarySearch(list_of_integers, target):
    success, count = _BinarySearch(list_of_integers, target, 0, len(list_of_integers))
    if not success:
        print(f"NOT FOUND \t{target}")
    return count
    
def _BinarySearch(list_of_integers, target, low, high):
    if high < low:
        return (False, 0)
    
    pivot = (low + high) // 2
    pivotValue = list_of_integers[pivot]
    if pivotValue == target:
        print(f"FOUND     \t{target}")
        return (True, 0)
    elif high <= low: # Must be 2-or-more element array
        return (False, 0)
    elif target > pivotValue: 
        success, count = _BinarySearch(list_of_integers, target, pivot+1, high)
        return (success, count+1)
    elif target < pivotValue:
        success, count = _BinarySearch(list_of_integers, target, low, pivot-1)
        return (success, count+1)
```

```python
# Tang Xuyuan

# Task 3.3

import random

# Read integers
file = open("SORTED_TangXuyuan_NUM_INDEX.txt", "r")
integers = []
for line in file:
    if line != "":
        integers.append(int(line))
file.close()

# Generate random integers
SAMPLESIZE = 50
randomIntegers = [random.randint(1, 200) for _ in range(SAMPLESIZE)]

# Run searches
totalComparisons = 0
for randomInt in randomIntegers:
    totalComparisons += BinarySearch(integers, randomInt)
    
# Display stats
avg = totalComparisons / SAMPLESIZE
print(f"Average comparisons over {SAMPLESIZE} samples: {avg}")
```
