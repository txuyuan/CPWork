#### Task 1

Write a function: **partition(A, low, high)** that accepts 3 arguments, array A, low index and high index, and returns the pivot position of the array.

* take the middle element of the array as a pivot
* find the position of the pivot:
    * swap the pivot value from the middle index to the low index of the array.
    * assign a left and right index to next index after the low index and high index respectively
    * move the left index to the right until a value greater than the pivot value is found
    * move the right index to the left until a value less than the pivot value is found
    * if left index is less than the right index, swap the their values
    * otherwise, swap the value of the right index with the value of the pivot 
    * return pivot index


Test your function with the following values:
```
A = [13, 12, 57, 99, 39, 64, 59, 48, 70]
low = 0
high = 8
```

````python
def partition(A, low, high):  
    pivot = (low+high)//2
    A[low], A[pivot] = A[pivot], A[low]
    p = A[low]
    left = low+1
    right = high
    
    while (left <= right):
        while (left <= high) and (A[left] < p):
            left += 1
        while (A[right] > p):
            right -= 1
        if (left <= right):
            A[left], A[right] = A[right], A[left]
    
    # now, right < pivot, left > pivot
    A[right], A[low] = A[low], A[right]
    return right

A = [13, 12, 57, 99, 39, 64, 59, 48, 70]
low = 0
high = 8
pivot = partition(A, low, high)
print(pivot)
````

Write a function **task5_1(S,n)** that accepts a string **S** of 10 characters and an integer **n** and returns an integer.

The function calculates an integer value using the following algorithm:
   
   * compute the sum of the products of each character's ASCII value and its corresponding index in the string  **S**.
   
   * the non-zero string indices commence from the rightmost character.
   
   * the sum is divided by **n** and the function returns the remainder.


````python
def task5_1(S, n):
    result = 0
    for i in range(len(S)):
        j = len(S) - i - 1
        result += ord(S[i]) * j
    return result % n

S = "abcdefghij"
n = 10
task5_1(S, n)
````

##### Task 5.3
Implement class WordIntPair with the following attributes and methods.
    
* Attributes:

    * word

    * number
 
 
* ``__str__(self)`` method that returns a string representation of the object with the following format:

    *  word, number
 
 
* ``compareWord(self, obj)`` method accepts another WordIntPair object and returns the follwoing:
    * 1 if its word is greater than
    * 0 if its word is equal
    * -1 if its word is less than 


* ``compareNum(self, obj)`` method accepts another WordIntPair object and returns the follwoing:
    * 1 if its number is greater than
    * 0 if its number is equal
    * -1 if its number is less than 
    

````python
class WordIntPair():
    def __init__(self, word, number):
        self._word = word
        self._number = number
    
    def get_word(self):
        return self._word
    
    def get_number(self):
        return self._number
    
    def __str__(self):
        return f"{self._word}, {self._number}"
    
    def compareWord(self, obj):
        value = obj.get_word()
        if value > self._word:
            return 1
        elif value < self._word:
            return -1
        else:
            return 0
        
    def compareNum(self, obj):
        value = obj.get_number()
        if value > self._number:
            return 1
        elif value < self._number:
            return -1
        else:
            return 0
````

#### Task 5.5
Write a function ```task5_5(A, sortBy,  asc)``` that accepts 3 arguments.

The function sorts the array of WordIntPair ```A``` using **insertion sort algorithm**. 

The argument ```sortBy``` allows 2 values, 'word' or 'number' which indicates which attribute 
the WordIntPir should be sorted.  

If ```asc``` is True, the function will sort array A in ascending order ,otherwise descending order. 

The argument ```sortBy``` has a default value ```'work'``` and  ```asc``` has default value of True.

````python
def task5_5(A, sortBy='word', asc=True):
    B = []
    for target in A:
        B = [target] + B # Append at beginning
        for i in range(len(B) - 1): # Bubble through once
            left, right = i, i+1
            
            asc_comparison = asc and (sortBy=='word' and B[right].compareWord(B[left])) or (sortBy=='number' and B[right].compareNum(B[left]))
            desc_comparison = not asc and (sortBy=='word' and not B[right].compareWord(B[left])) or (sortBy=='number' and not B[right].compareNum(B[left]))
            
            if (asc_comparison) or (desc_comparison):   
                B[left], B[right] = B[right], B[left]
    A = B
    return A
````

````python
objects = task5_4("assets/task5_2.txt")
task5_5(objects, "number", False)
for x in objects:
    print(x)
````
