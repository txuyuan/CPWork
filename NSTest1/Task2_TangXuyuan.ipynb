```python
# Task 2.1

import sqlite3

DB_FILE = "bakery.db"
CAKE_FILE = "CAKES.txt"
LOAVES_FILE = "LOAVES.txt"
BUNS_FILE = "BUNS.txt"

def task2_1():
    con = sqlite3.connect(DB_FILE)
    con.row_factory = sqlite3.Row
    
    for l in open(CAKE_FILE, 'r'):
        add_cake(l.strip().split(","), con)
        
    for l in open(LOAVES_FILE, 'r'):
        add_loaf(l.strip().split(","), con)
        
    for l in open(BUNS_FILE, 'r'):
        add_bun(l.strip().split(","), con)
        
def add_cake(cake, con):
    product = cake[0:2] + ["Cake"] + cake[2:4]
    cake = cake[0:1] + cake[4:6]
    
    cur = con.cursor()
    cur.execute("""
        INSERT INTO Cake(ProductCode, ServingSize, Shape) VALUES(?, ?, ?)
    """, tuple(cake))
    con.commit()
    
    add_product(product, con)
    
def add_loaf(loaf, con):
    product = loaf[0:2] + ["Loaf"] + loaf[2:4]
    loaf = loaf[0:1] + loaf[4:5]
    
    cur = con.cursor()
    cur.execute("""
        INSERT INTO Loaf(ProductCode, Weight) VALUES(?, ?)
    """, tuple(loaf))
    con.commit()
    
    add_product(product, con)

def add_bun(bun, con):
    product = bun[0:2] + ["Bun"] + bun[2:4]
    bun = bun[0:1] + bun[4:5]
    
    cur = con.cursor()
    cur.execute("""
        INSERT INTO Bun(ProductCode, PiecesPerPackage) VALUES(?, ?)
    """, tuple(bun))
    con.commit()
    
    add_product(product, con)
    

def add_product(product, con):
    cur = con.cursor()
    cur.execute("""
        INSERT INTO Product(ProductCode, Name, Type, Location, Price) VALUES(?, ?, ?, ?, ?)
    """, tuple(product))
    con.commit()
    print("Committed " + product[0])

task2_1()
```

```python
# Task 2.2

import sqlite3

DB_FILE = "bakery.db"
PRODUCTS = ["Cake", "Loaf", "Bun"]
LOCATIONS = ["North", "South", "East", "West"]

PRODUCT_DETAILS = {
    "Cake": ["ServingSize", "Shape"],
    "Loaf": ["Weight"],
    "Bun": ["PiecesPerPackage"]
}

def task2_2():
    success1, success2 = False, False
    while(not success1):
        product_type = input("Enter Product Type: ")
        if product_type in PRODUCTS:
            success1 = True
        else:
            print(f"Invalid type, please enter valid type: {PRODUCTS}")
    while(not success2):
        location = input("Enter Location: ")
        if location in LOCATIONS:
            success2 = True
        else:
            print(f"Invalid type, please enter valid type: {LOCATIONS}")
        
    con = sqlite3.connect(DB_FILE)
    cur = con.cursor()
    
    specific_parameters = ', '.join(PRODUCT_DETAILS[product_type])
    cur.execute(f"""
        SELECT Location, Product.ProductCode, Name, Price, {specific_parameters}
        FROM {product_type}
        INNER JOIN Product ON {product_type}.ProductCode = Product.ProductCode
        WHERE Location = ?
    """, (location,))
    products = cur.fetchall()
    
    for p in products:
        print("> " + p.__str__().strip('()'))
    
task2_2()
```
