### Task 1.1
The file `INTEGERS.txt` stores 100 integers. Write a program to read the integers, arrange them in ascending order using quick sort, and write the sorted integers to a file called

    SORTED_<your name>_<centre number>_<index number>.txt [15]

```python
# Task 1.1

def partition(A, low, high):
    pivot = low
    left = low+1
    right = high

    while right >= left:
        while right >= left and A[left] <= A[pivot]:
            left += 1
        while right >= left and A[right] > A[pivot]:
            right -= 1
        if right >= left:
            A[left], A[right] = A[right], A[left]
    A[pivot], A[right] = A[right], A[pivot]
    pivot = right
    return pivot

def quicksort(A, low=0, high=None):
    if high==None:
        high = len(A)-1
    if high <= low:
        return
    
    pivot = partition(A, low, high)
    quicksort(A, low, pivot-1)
    quicksort(A, pivot+1, high)

integers = []
with open("INTEGERS.txt", "r") as file:
    # File auto closes
    for line in file:
        integers += [int(line.strip())]
quicksort(integers)
print(integers)

with open("SORTED_TangXuyuan_3036_0403.txt", "w+") as file:
    # File closes automatically
    integer_str = [str(x) for x in integers]
    file.write("\n".join(integer_str))
```

#### Task 1.2
Write a function `BinarySearch(list_of_integers, target)` that

- takes a list of ascending integers, `list_of_integers` and an integer `target`
- performs a binary search
- prints out if `target` is found in `list_of_integers`
- returns the number of comparisons during the binary search 
                                                                                        
                                                                    [8]

```python
# Task 1.2

def BinarySearch(list_of_integers, target):
    left, right = 0, len(list_of_integers)-1
    comparisons = 0
    while right > left:
        comparisons += 1
        pivot = left + (right-left) // 2
        if list_of_integers[pivot] == target:
            left = right = pivot
        elif list_of_integers[pivot] < target:
            left = pivot + 1
        else:
            right = pivot - 1

    if left == right and list_of_integers[left] == target:
        print("Found!")
    else:
        print("Not found")

    return comparisons  
```

### Task 3
The task is to implement a priority queue using Object-Oriented Programming.

A priority queue is an extension of queue with the following properties.
- Every element has a priority associated with it. Smaller integer value has a higher priority.
- An element with high priority leaves the queue before an element with low priority.
- If two elements have the same priority, they are served according to their order in the queue, i.e. the earlier element will be served before the later element (FIFO).

For example, the emergency room in a hospital assigns patients with priority numbers. The patient with the highest priority is treated first, regardless of the order of arrival.

`Refer to PDF for example.`

A priority queue abstract data type (ADT) is to be implemented as a linked list using objectoriented programming. Two classes Node and PQueue have been identified.

                                   Class: Node

| Identifier  | Data Type  | Description  |
|:-|:-|:-|
| Data         |   STRING      |   The node data
| Priority     |   INTEGER     |   Indicates priority of node. Smaller value has higher priority
|  Next         |   INTEGER     |   Pointer to next node in queue.
    
    
    Class: PQueue

| Identifier | Data Type | Description |
|:-|:-|:-|
| ThisPQueue | ARRAY[1:10] Of Node | The priority queue data. |
| Front | INTEGER | Index for front node of queue. |
| Rear | INTEGER | Index for rear node of queue. |
| NextFree | INTEGER | Index for the next unused node. |
| Initialise() | PROCEDURE| ● Set pointers to indicate all nodes are unused and linked. ● Initialise values for Front, Rear and NextFree.|
| PQInsert(NewItem, Priority) | PROCEDURE | ● Assign NewItem and Priority passed as parameters to a node. ● Insert the node to the rear of the priority queue. |
| PQDelete() | FUNCTION | ● Remove a node of highest priority from the priority queue. ● Return the Data attribute of the node that is removed.|
| DisplayPQueue() | PROCEDURE | ● Display the values of Front,Rear, NextFree and the contents of ThisPQueue array in index order. |

#### Task 3.1
Write program code for the classes Node and PQueue, including the Initialise, PQInsert, PQDelete and DislayPQueue methods. The code should follow the specification given.   [17]

```python
# Task 3.1

class Node():
    def __init__(self, data=None, priority=-1, nxt=-1):
        self._data = data
        self._priority = priority
        self._nxt = nxt
        
    def get_data(self):
        return self._data
    def set_data(self, data):
        self._data = data
    def get_priority(self):
        return self._priority
    def set_priority(self, priority):
        self._priority = priority
    def get_nxt(self):
        return self._nxt
    def set_nxt(self, nxt):
        self._nxt = nxt

class PQueue():
    def Initialise(self):
        # Initialise node array
        queue = []
        for i in range(10):
            if i < 10-1:
                queue += [Node(nxt=i+1)] # Leave data and priority to uninitialised null values
            else:
                queue += [Node(nxt=-1)] # Set last element to point to -1
        self._ThisPQueue = queue

        # Initialise front & rear to null values
        self._Front = -1
        self._Rear = -1

        self._NextFree = 0

    def PQInsert(self, NewItem, Priority):
        if self._NextFree == -1:
            print("List full!")
            return

        # Insert values into free node
        new = self._NextFree
        new_node = self._ThisPQueue[new]
        self._NextFree = self._ThisPQueue[self._NextFree].get_nxt()
        new_node.set_data(NewItem)
        new_node.set_priority(Priority)
        new_node.set_nxt(-1)

        # Update front and rear
        if self._Front == -1: 
            # Empty state case
            self._Rear = self._Front = new
        else:
            self._ThisPQueue[self._Rear].set_nxt(new)
            self._Rear = new

    def PQDelete(self):
        # Empty case
        if self._Front == -1:
            print("List empty!")
            return

        # Find highest priority node
        high_priority = -1
        # To facilitate deletion, store current highest node and node directly prior
        high_prev, high = -1, -1

        prev, probe = -1, self._Front
        while self._ThisPQueue[probe].get_nxt() != -1:
            if high_priority == -1 or self._ThisPQueue[probe].get_priority() < high_priority:
                high_priority = self._ThisPQueue[probe].get_priority()
                high_prev, high = prev, probe
                
            prev, probe = probe, self._ThisPQueue[probe].get_nxt()

        # Retrieve target data
        data = self._ThisPQueue[high].get_data()

        if prev == -1:
            # Queue only has one element, and the probe never advanced => Must reset to empty state
            self._Front = self._Rear = -1

        elif high_prev == -1:
            self._Front = self._ThisPQueue[high].get_nxt()
            # First element is selected
            
        elif self._ThisPQueue[high].get_nxt() == -1:
            # Last element is selected
            self._Rear = high_prev
            
        else:
            # Delete
            skip = self._ThisPQueue[high].get_nxt()
            self._ThisPQueue[high_prev].set_nxt(skip)

        return data

    def DisplayPQueue(self):
        print(f"Front: {self._Front}")
        print(f"Rear: {self._Rear}")
        for i, node in enumerate(self._ThisPQueue):
            print(f"[{i}] Data: {node.get_data()}, priority: {node.get_priority()}, next: {node.get_nxt()}")
```

The task is to write program code for a Tic-Tac-Toe-Tomek game for two players.

Tic-Tac-Toe-Tomek is a game played on a 4 x 4 square board. The board starts empty, except that a single 'T' symbol may appear in one of the 16 squares. There are two players: X and O. They take turns to make moves, with X starting. In each move a player puts her symbol in one of the empty squares. Player X's symbol is 'X', and player O's symbol is 'O'.

After a player's move, if there is a row, column or a diagonal containing 4 of that player's symbols, or containing 3 of her symbols and the 'T' symbol, she wins and the game ends. Otherwise the game continues with the other player's move. If all of the fields are filled with symbols and nobody won, the game ends in a draw.

Given the 4 x 4 board description containing 'X', 'O', 'T' and '.' characters (where '.' represents an empty square). The following examples show the various winning positions.

Write a function displayBoard that will display the game board clearly to the players. You should use the board as a parameter in displayBoard. [2]

```python
# Task 4.2

def displayBoard(board):
    print()
    print("  A B C D")
    for i, line in enumerate(board):
        print(str(i+1) + " " + " ".join(line))
    print()
```

#### Task 4.3
Write a function getPlayerMove to get players to make their move (by marking 'X' or 'O') on the board. You should include validation on player’s input and check that the space is not already occupied. Use board as a parameter. You may include any other suitable parameters.                                                   [4] 

```python
# Task 4.3

import re

def getPlayerMove(board, player):
    accepting = True
    while accepting:
        inp = input(f"Player {player}: Please enter desired coordinate to mark [A1, etc]: ")
        if len(inp) != 2:
            print("Invalid length")
        elif not re.search("[A-D][1-4]", inp):
            print("Invalid characters")
        else:
            x = ord(inp[0]) - 65
            y = int(inp[1]) - 1

            if board[y][x] != ".":
                print("Space already taken!")
            else:
                board[y][x] = player
                accepting = False
```

#### Task 4.4
Write a function checkWin that checks all the conditions for winning a game and returns True if a player has won the game, otherwise returns False. Use board as a parameter. You may include any other suitable parameters. [5]

```python
# Task 4.4

def checkMask(board, mask_indices, player):
    win = True
    for i in mask_indices:
        win = win and (board[i] == player or board[i] == "T")
    return win

def checkWin(board):
    # Collate all winning patterns
    winning_masks = []
    for row_left in range(0, 15, 4):
        winning_masks += [list(range(row_left, row_left + 4))]
    for col_top in range(4):
        winning_masks += [[col_top + i*4 for i in range(4)]]
    winning_masks += [[i*5 for i in range(4)]]
    winning_masks += [[(i+1)*3 for i in range(4)]]

    # Delineate board into 1D array
    board_map = []
    for line in board:
        board_map += line

    # For each player, check for all winning patterns
    for i, player in enumerate(["X", "O"]):
        won = False
        for mask in winning_masks:
            won = won or checkMask(board_map, mask, player)
        if won:
            return (True, player)

    # Special draw case: all cells are filled:
    draw = True
    for i in range(16):
        draw = draw and board_map[i] != "."
    if draw:
        return (True, "D")

    # Final default condition: no one has won
    return (False, None)
```
