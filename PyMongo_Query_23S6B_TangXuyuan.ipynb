````python
import pymongo
from pymongo import MongoClient

client = MongoClient("localhost", 27017)

bookdb = client['bookdb']
books = bookdb['book']

query = {'pageCount': 
         {"$gt": 600},
        }
projection = {'title': 1, 'pageCount': 1}
result = books.find(query, projection).sort("pageCount")

print(f"{'Title':<60} Page")
for book in result:
    print(f"{book['title']:<60} {book['pageCount']}")
    

client.close()
````

Find books with categories contains both **Java** and **Internet**. Display the first 10 documents as shown.

````python
import pymongo
from pymongo import MongoClient

client = MongoClient("localhost", 27017)

bookdb = client['bookdb']
books = bookdb['book']

query = {
          'categories': {
              "$all": ["Java", "Internet"]
          }
        }
projection = {'title': 1, 'categories': 1}
result = books.find(query, projection).limit(10).sort("title")

print(f"{'Title':<60} Categories")
for book in result:
    print(f"{book['title']:<60} {', '.join(book['categories'])}")
    

client.close()
````

````python
import pymongo
from pymongo import MongoClient

client = MongoClient("localhost", 27017)

bookdb = client['bookdb']
books = bookdb['book']

query = {
          'categories': "Java",
          'authors': {
              "$exists": True
          },
          'pageCount': {
              "$gt": 550
          }
        }
projection = {'title': 1, 'pageCount': 1, 'categories': 1, 'authors': 1}
result = books.find(query, projection).sort("title").sort("pageCount", -1)

print(f"{'Title':<45} {'Page Count':<15} {'Categories':<20} Author/s")
for book in result:
    print(f"{book['title']:<45} {book['pageCount']:<15} {', '.join(book['categories']):<20} {', '.join(book['authors'])}")
    

client.close()
````
