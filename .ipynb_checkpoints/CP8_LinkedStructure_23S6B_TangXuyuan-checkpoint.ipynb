### Coding Practice 8 - Singly Linked List

###### Task 1

Implement the class Node which has the following attributes:
* `_data`
* `_next`

and methods:
* `getData()` returns _data of the Node object.
* `getNext()` returns _next of the Node object.
* `setData()` set the value of _data of the Node object.
* `setNext()`  set the value of _next of the Node object.
* `__str__`  returns the string representation of the Node object.
* `__init__`  accepts two arguments 'data', 'nxt' and initialize the Node object with the value passed in.

Write codes to test the Node class.

```python
class Node:
    def __init__(self, data, nxt=None):
        self._data = data
        self._next = nxt
    def getData(self):
        return self._data
    def getNext(self):
        return self._next
    def setData(self, data):
        self._data = data
    def setNext(self, nxt):
        self._next = nxt
    def __str__(self):
        return f"> {self.getData()} > {self.getNext().getData()}"
    def __str_all__(self):
        if self.getNext() == None:
            return f"> {self.getData()}"
        else:
            return f"> {self.getData()}\n{self.getNext().__str_all__()}"
```

###### Task 2

Implement the class `SinglyLinkedList` which has the following attributes:
* `_head`

and methods:
* `getHead()` returns _head of the SinglyLinkedList object.
* `setHead()` set the value of _head of the SinglyLinkedList object.
* `__str__()` returns the string representation of the SinglyLinkedList object.
* `__init__()` initializes the SinglyLinkedList object to an empty list.
* `search()` accepts an argument: `target`. If `target` is found, returns the position , otherwise returns -1.
* `replace()` accepts two arguments: `old`, `new`. Replace the first node with `old` data value with `new` data value.
* `replaceAll()` accepts two arguments: `old`, `new`. Replace all nodes with `old` data value with `new` data value.
* `replaceByPosition()` accepts two arguments: `pos`, `new`. Replace all nodes with `old` data value with `new` data value.
* `isEmpty()` returns True if list is empty, otherwise, False.
* `size()` returns the number of nodes in the list.
* `insert()` accepts two arguments: `pos`, `node`. Insert `node` at the position indicated by `pos`. Assume the first node is at position zero. If `pos`> length-1, append `node`  
* `remove()` accepts an arguments: `pos`. Remove `node` at the position indicated by `pos`. If `pos`> length-1, remove the last `node` 
* `append()` accepts an arguments: `node`. Append `node` at the end of the list.

Write codes to test the SinglyLinkedList class.

```python
class SinglyLinkedList:
    def __init__(self):
        self._head = None
        
    def getHead(self):
        return self._head
    def setHead(self, head):
        self._head = head
    
    def __str__(self):
        return f"List: \n{self._head.__str_all__()}"
    
    def search(self, target):
        probe = self._head
        index = 0
        while (probe != None):
            if probe.getData() == target:
                return index
            probe = probe.getNext()
            index += 1
        return -1
    
    def replace(self, old, new):
        probe = self._head
        while (probe != None):
            if probe.getData() == old:
                probe.setData(new)
                return
            probe = probe.getNext()
                
    def replaceAll(self, old, new):
        probe = self._head
        while(probe != None):
            if probe.getData() == old:
                probe.setData(new)
            probe = probe.getNext()
                
    def replaceByPosition(self, pos, new):
        probe = self._head
        index = 0
        while(index < pos):
            if probe==None:
                return
            probe = probe.getNext()
            index += 1
        probe.setData(new)
        
    def insert(self, pos, node):
        if pos > self.size()-1:
            self.append(node)
            
        probe = self._head
        index = 0
        while (index+1 < pos):
            if probe==None:
                return
            probe = probe.getNext()
            index += 1
        node.setNext(probe.getNext())
        probe.setNext(node)
        
    def remove(self, pos):
        if pos == 0:
            self._head = self._head.getNext()
            return
        
        probe = self._head
        index = 0
        while (index+1 < pos):
            if probe==None:
                return
            probe = probe.getNext()
            index += 1
        probe.setNext(probe.getNext().getNext())
            
        
    def append(self, node):
        if self._head == None:
            self._head = node
            return
        
        probe = self._head
        while (probe.getNext() != None):
            probe = probe.getNext()
        probe.setNext(node)
        
        
    def isEmpty(self):
        return self._head == None or self._head.getData() == None      
    def size(self):
        probe = self._head
        index = 0
        while (probe != None):
            probe = probe.getNext()
            index += 1
        return index
    

l = SinglyLinkedList()
for i in range(65, 84):
    l.append(Node(chr(i), None))
    l.append(Node(chr(i), None))
    
print(l)

print(f"\nSearch > {l.search('F')}\n")

l.replace("A", "AAA") # Expect AAA, A,

l.replaceAll("B", "BBB") # Expect AAA, A, BBB, BBB

l.replaceByPosition(4, "CCC") # ExpectA CCC, C

l.insert(6, Node("DDD", None)) # Expect

l.remove(9)

l.append(Node("Z", None))

print(f"IsEmpty: {l.isEmpty()}")
print(f"Size: {l.size()}")

print(l)
```

###### Task 4

Using the Singly Linked List created in `Task3`. 
Write code to 
* shift all nodes that contain odd number to the end of the list. This can be achieved by
    * Searching all odd number node in `lst` 
    * Removing the node with odd number 
    * Create a new node with the odd number and append it to the end of the list
* print out the content of the `lst`

```python
### Task 4
### Your code here!

def oddToBack(lst):
    for i in range(1, 100, 2):
        index, counter, size = lst.search(i), 0, lst.size()
        while (index != -1 and counter < size):
            lst.remove(index)
            lst.append(Node(i, None))
            index = lst.search(i)
            counter += 1

oddToBack(lst)
print(lst)
```

###### Task 5

Using the Singly Linked List in `Task4`. 
Write code to 
* create a node for each even number node in `lst` and append the new node to the end of the lst.
    * This can be achieved by searching all even number node in `lst` 
    * When found, create a new node with the even number and appending it to the end of the list
* print out the content of the `lst`


```python
### Task 5
### Your code here!

def evenToBack(lst):
    for i in range(0, 100, 2):
        movedItems, size = 0, lst.size()
        
        lastIndex, index = -1, lst.search(i)
        while (index != -1 and index < size-movedItems and index!=lastIndex):
            lst.append(Node(i, None))
            lastIndex, index = index, lst.search(i)
            movedItems += 1

evenToBack(lst)
print(lst)
```

*For Task 6, methods should be implemented with either changing the links or using existing methods.

Implement the class `ShiokLinkedList` which extends `SinglyLinkedList` which has the following methods:

* `moveOddNodeToBack()` traverse the list and move nodes that contains odd number in data to the end of the list.
* `doubleEvenNode()` traverse the list, when a node that contains even number is found, create a node with the same value and add the node to the front of the list.
* `reverse()` reverse the link list non recursively
* `reverse2()` reverse the link list recursively (if you can)
* `insertByValue()` accepts an argument `orderby` and the following applies:
    * `orderby` has a default value of `ASC`  
    * `orderby` strictly accepts `ASC` and `DESC`
    * insert `node` at the correct position such that all nodes are arranged in ascending order when `orderby` is `ASC`.
    * insert `node` at the correct position such that all nodes are arranged in descending order when `orderby` is `DESC`
    




```python
### Task 6
### Your code here!

class ShiokLinkedList(SinglyLinkedList):
    def __init__(self):
        super().__init__()
        
    def moveOddNodeToBack(self):
        size = self.size()
        probe = Node(0, self._head)
        i, j = 0, -1
        while (probe.getNext() != None and i < size):
            if j == 0:
                self._head = probe
            
            if probe.getNext().getData() % 2 == 1:
                # Odd
                self.append(Node(probe.getNext().getData()))
                probe.setNext(probe.getNext().getNext())
            else:
                # Even
                probe = probe.getNext()
                j += 1
            i += 1
        if j == -1:
            self._head = probe.getNext()
    
    def doubleEvenNode(self):
        size = self.size()
        probe = Node(1, self._head)
        i = 0
        while (probe.getNext() != None and i < size):
            if probe.getNext().getData() % 2 == 0:
                self.append(Node(probe.getNext().getData()))
            probe = probe.getNext()
            i += 1
            
    def reverse(self):
        ptr1, ptr2, ptr3 = self._head, self._head.getNext(), self._head.getNext().getNext()
        ptr1.setNext(None)
        if (self.size()==2):
            # Take care of edge case
            ptr2.setNext(ptr1)
        while (ptr3 != None):
            ptr2.setNext(ptr1)
            ptr1, ptr2, ptr3 = ptr2, ptr3, ptr3.getNext()
        self.setHead(ptr2)
        
    def reverse2(self):
        if self.size()==1:
            return
        
        data = self._head.getData()
        self.remove(0)
        self.reverse2()
        self.append(Node(data))
        
    def insertByValue(self, node, orderby="ASC"):
        probe = self._head
        pos = 0
        while((orderby=="ASC" and node.getData()<probe.getData()) or (orderby=="DESC" and node.getData()>probe.getData())):
            probe = probe.getNext()
            pos += 1
        self.insert(pos, node)

def generateList():
    lst = ShiokLinkedList()
    for i in range(10):
        num = random.randint(1, 100)
        node = Node(num)
        lst.append(node)
    return lst
      
print("\nOdd Node")
lst=generateList()
print(lst)
lst.moveOddNodeToBack()
print(lst)

print("\nEven Node")
lst=generateList()
print(lst)
lst.doubleEvenNode()
print(lst)

print("\nReverse Node")
lst=generateList()
print(lst)
lst.reverse()
print(lst)

print("\nRecursive Reverse Node")
lst=generateList()
print(lst)
lst.reverse2()
print(lst)

print("\nInsert ASC Node")
lst=generateList()
print(lst)
lst.insertByValue(Node(50))
print(lst)

print("\nInsert DESC Node")
lst=generateList()
print(lst)
lst.insertByValue(Node(50), "DESC")
print(lst)
```
